```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data using pandas

dataset = pd.read_csv("/content/sample_data/GOOG.csv")
```

```python
#building the model

#select features: date and price

data = dataset[["Date", "Close"]]


#rename colums 
data = data.rename(columns = {"Date":"ds", "Close":"y"})

data.head()
```

```python
#install facebook Fbprothet

!pip install fbprophet
```

```python
#import model

from fbprophet import Prophet

model = Prophet(daily_seasonality=True)

```

```python
#fit the model on all of the data

model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365) #predit 1 year 

```

```python
prediction = model.predict(future)

model.plot(prediction)

plt.title("Prediction of Prices")
plt.xlabel("Date")
plt.ylabel("Close Stock Price")
plt.show()
```

```python
model.plot_components(prediction)
plt.show()
```

```python
apple_dataset = pd.read_csv("/content/sample_data/AAPL.csv")
```

```python
apple_data = apple_dataset[["Date", "Close"]]

apple_data = apple_data.rename(columns = {"Date":"ds", "Close":"y"})
```

```python
apple_model = Prophet(mcmc_samples=2300)

```

```python
apple_model.fit(apple_data)
```

```python
future = apple_model.make_future_dataframe(periods=365) #predit 1 year 
```

```python
prediction = apple_model.predict(future)

apple_model.plot(prediction)

plt.title("Prediction of Prices")
plt.xlabel("Date")
plt.ylabel("Close Stock Price")
plt.show()
```

```python
apple_model.plot_components(prediction)
plt.show()
```
